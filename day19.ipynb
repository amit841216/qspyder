which is used to calculate the num or factor except itself 

```python
num = int(input("enter a num:"))
i = 1
sum = 0
while i<num:
    if num%i == 0:
        sum +=i
    i+=1
if sum == num:
    print("perfect")
else:
    print("not perfect")
```
