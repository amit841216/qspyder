```python
def uper_case(st):
    out = " "
    for i in st:
        if 'a' <=i <='z':
            out +=chr(ord(i )-32)
        elif 'A' <= i <= 'Z':
            out += chr(ord(i)+32)
    print(out)
        





st = 'Dhiraj'
uper_case(st)
```

```python
a = ['hello', 10,4.8,[1,2],3+5j]
def len_of_a(a):
    out = []
    for i in a:
        if type(i)in[str,tuple,set,dict  ]:
            out+=[(i,len(i))]
        else:
            out+=[(i,1)]
    print(out)
        


len_of_a(a)
```
