```python
a = 'Ansh Singh'
out = ' '
for i in a:
    if 'a' <= i <= 'z':
        continue 
    else:
        out+=i   
print(out)
```

extract int from a given list but it shoud be even index and palandrom

```python
a = [212,322,455,67,787,657,453,345,545]
out = []
for i in range(len(a)):
    if type(a[i])==int and i%2==0 and str(a[i   ]) == str(a[i])[::-1]:
        out.append(a[i])

print(out)
```
